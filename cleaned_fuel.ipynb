# Import packages required (pandas, matplotlib, and seaborn), read in the csv file, and view the first 5 rows.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='dark')
df = pd.read_csv('fuel.csv')
df.head()
```

# Convert the column names to lowercase, rename the trans column to 'transmission', and replace whitespaces in the column names.

```python
# Converts the column names to lowercase.
df.columns = df.columns.str.lower()

# Renames the 'trans' column to 'transmission'.
df.rename(columns = {'trans':"transmission"}, inplace=True)

# Replaces whitespace in the column names (if there are any) with nothing.
df.columns.str.replace(' ', '')

df.head()
```

# Splitting and dropping unwanted columns.

```python
# Creates two columns to hold the values of the transmission column after splitting.
df[['tranmission', 'transmission_1']] = df['transmission'].str.split(" ", n=1, expand=True)
```

```python
df.head()
```

```python
# Dropping the first 'transmission' and the 'transmission_1' columns.
df.drop(["transmission", "transmission_1"], axis=1, inplace=True)
```

```python
df.head()
```

# Visualising the data.

### 1. Univariate Visual - Categorical

```python
sns.countplot(data=df, y='vclass', color='blue')
plt.title('Cars by Class')
plt.xlabel('Number of Cars')
plt.ylabel('Car type');
```

### 2. Univariate Visual - Quantitative

```python
sns.histplot(data=df, x='co2 emission', color='blue')
plt.title('Amount of CO2 Emitted')
plt.xlabel('CO2 Emissions')
plt.ylabel('Number of Cars')
```

### 3. Bivariate Visual: 1 Categorical, and 1 Quantitative.

```python
sns.violinplot(data=df, x='year', y='co2 emission', color='orange')
plt.title('CO2 Emissions by Year')
plt.xlabel('Year')
plt.ylabel('Amount of CO2 Emitted');
```

### 4. Bivariate Visual: 2 Quantitatives.

```python
sns.scatterplot(data=df, y='drive', x='cylinders', color='orange')
plt.title('Correlation between Drive System and Cylinders')
plt.xlabel('Cylinder Type')
plt.ylabel('Drive System')
plt.text(14, 5, 'Chart shows no correlation between Drive System and Cylinder types.');
```

### 5. Bivariate Visuals: 2 Categoricals.

```python
sns.barplot(data=df, x='drive', y='co2 emission', hue='tranmission', ci=None)
plt.title('CO2 Emission by Transmission Type on Drive Sytem')
plt.xlabel('Drive Sytem')
plt.ylabel('CO2 Emission')
plt.xticks(rotation=45, horizontalalignment='right');
```
